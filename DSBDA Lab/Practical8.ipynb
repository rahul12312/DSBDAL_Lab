```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Calculate the correlation matrix for the numerical columns
corr = tips.corr()

# Create the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', cbar=True)

# Show the plot
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', color='blue')
plt.show()
```

```python


# Create a simple pairplot (without extra customizations)
sns.pairplot(tips)

# Show the plot
plt.show()
```

```python
tips = sns.load_dataset('tips')
sns.displot(tips['total_bill'], kde=True)
plt.show()
```

```python
sns.countplot(x='sex', data=tips)
plt.show()
```

```python
sns.boxplot(x='time', y='total_bill', data=tips)
plt.show()
```

```python
sns.boxplot(x='time', y='total_bill', data=tips, palette={"Lunch": "#FF6347", "Dinner": "#4682B4"})
plt.show()
```

```python
sns.violinplot(x='time', y='total_bill', data=tips, palette={"Lunch": "#FF6347", "Dinner": "#4682B4"})
plt.show()
```
